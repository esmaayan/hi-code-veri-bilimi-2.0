.plot.barh() = verileri yatay bir çubuk grafik (bar plot) olarak görselleştirir.

```python
df_dropped["JoiningYear"].value_counts().plot.barh()
```

```python
df_dropped["JoiningYear"].value_counts().plot.bar()
```

```python
joinYear = df_dropped["JoiningYear"].value_counts().sort_index()
plt.bar( x = joinYear.index, height = joinYear.values)
```

```python
plt.figure(figsize=(10,5))

plt.bar(x = joinYear.index, height= joinYear.values)
plt.xlabel("year")
plt.ylabel("count")
plt.title("count of employee by year")
```

```python
plt.figure(figsize=(10,5))

sns.barplot(x = joinYear.index, y= joinYear.values, palette="Set3")
plt.xlabel("year")
plt.ylabel("count")
plt.title("count of employee by year")
```
